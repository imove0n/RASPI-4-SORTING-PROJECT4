## 📤 CELL 2: Upload Your 40 Tomato Images

**Click the ▶ play button**

A file upload dialog will appear. Select all 40 images.

```python
from google.colab import files
import os
import shutil

print("📤 Upload your 40 tomato images")
print("Click 'Choose Files' and select all 40 images\n")

# Create directory for images
os.makedirs('/content/tomato_images', exist_ok=True)

# Upload images
uploaded = files.upload()

# Move uploaded files
for filename in uploaded.keys():
    shutil.move(filename, f'/content/tomato_images/{filename}')

# Count images
image_count = len(os.listdir('/content/tomato_images'))

print(f"\n✅ Uploaded {image_count} images successfully!")
print("\n👉 Next: Run CELL 3 to start Label Studio")
```

## 🚀 CELL 3: Start Label Studio

**Click the ▶ play button**

This will:
1. Start Label Studio server
2. Create labeling project
3. Give you a clickable link

**Click the ngrok link** when it appears to open Label Studio!

```python
print("🚀 Starting Label Studio...\n")

# Install ngrok for public URL
!pip install -q pyngrok
from pyngrok import ngrok

# Set ngrok auth token (optional but recommended)
print("📝 Setting up Label Studio project...\n")

# Initialize Label Studio project
!label-studio init tomato_project \
  --input-path=/content/tomato_images \
  --input-format=image-dir \
  --label-config=/content/label_config.xml \
  --force

# Create label configuration
label_config = '''<View>
  <Image name="image" value="$image"/>
  <RectangleLabels name="label" toName="image">
    <Label value="ripe" background="#FF0000"/>
    <Label value="unripe" background="#00FF00"/>
  </RectangleLabels>
</View>'''

with open('/content/label_config.xml', 'w') as f:
    f.write(label_config)

print("✅ Project created!")
print("\n🌐 Starting server with public URL...\n")

# Start ngrok tunnel
public_url = ngrok.connect(8080)

print("="*60)
print("🎉 LABEL STUDIO IS READY!")
print("="*60)
print(f"\n🔗 Click this link to open Label Studio:")
print(f"\n   {public_url}\n")
print("="*60)
print("\n📋 INSTRUCTIONS:")
print("   1. Click the link above")
print("   2. Sign up or skip")
print("   3. Start labeling your tomatoes!")
print("   4. Draw boxes and select 'ripe' or 'unripe'")
print("   5. When done, run CELL 4 to export")
print("\n⚠️  Keep this cell running! Don't stop it!")
print("="*60)

# Start Label Studio
!label-studio start tomato_project --port 8080
```

## 💾 CELL 4: Export Your Labels

**After you finish labeling all images:**

1. Stop CELL 3 (click ⏹ stop button)
2. Run this cell to export labels in YOLO format
3. Download the ZIP file

```python
import json
import zipfile
from google.colab import files

print("💾 Exporting labels in YOLO format...\n")

# Export from Label Studio
!label-studio-converter export \
  -i /content/tomato_project/export.json \
  -o /content/yolo_dataset \
  --format YOLO

# Create classes.txt
with open('/content/yolo_dataset/classes.txt', 'w') as f:
    f.write('ripe\nunripe')

# Zip everything
print("📦 Creating ZIP file...\n")
shutil.make_archive('/content/tomato_labels', 'zip', '/content/yolo_dataset')

print("✅ Labels exported successfully!")
print("\n📥 Downloading ZIP file...\n")

# Download
files.download('/content/tomato_labels.zip')

print("="*60)
print("🎉 DONE!")
print("="*60)
print("\nYou now have:")
print("  ✅ tomato_labels.zip (contains all labels)")
print("\n👉 Next: Tell me 'Labels ready for training!'")
print("="*60)
```
